```python
import pandas as pd
import numpy as np
import re
```

```python
reviews = [
    "OMG!!! This movie was AMAZING 😍😍😍 Best film I've seen in 2024!!!",
    "Waste of time... wouldn't recommend to anyone. Rating: 2/10",
    "The acting was great, but the plot was sooooo boring 😴",
    "https://movie-reviews.com - Check out my full review here! #MustWatch",
    "I LOVED it! Tom's performance >>> everyone else's. Cost me $15 but worth it 💯",
    "   meh. it was okay i guess. not bad not good   ",
    "DON'T WASTE YOUR MONEY!!! Terrible acting & poor direction 👎👎",
    "The cinematography was beautiful. Released on 25th Dec, 2023. 9/10 stars ⭐",
    "idk why people hate this... I thought it was pretty good tbh 🤷",
    "Characters weren't well-developed. The storyline didn't make sense either."
]
```

## Converting to Lowercase

```python
for i in range(len(reviews)):
    reviews[i] = reviews[i].lower()
print(reviews)
```

```python
cleaned_reviews1 = reviews
```

## Removing URL's

```python
def remove_url(l):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return [pattern.sub(r'', i) for i in l]
```

```python
cleaned_reviews2 = remove_url(cleaned_reviews1)
```

```python
cleaned_reviews2
```

## Removing Emojis

```python
def remove_emoji(li):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return [emoji_pattern.sub(r'', i) for i in li]
```

```python
cleaned_reviews3 = remove_emoji(cleaned_reviews2)
```

```python
cleaned_reviews3
```

```python
import emoji
```

```python
def remove_all_emoji(text):
    return emoji.replace_emoji(text, replace='')
```

```python
remove_all_emoji(cleaned_reviews3)
```

```python
import string
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(lis):
    for char in exclude:
        lis = lis.replace(char, '')
    return lis
```

```python
cleaned_reviews4 = [remove_punc(i) for i in cleaned_reviews3]
```
